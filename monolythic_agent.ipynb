### I'll build an agent that can take the action of searching Google. The problem with the free use of gemini is that RPM using the google search is being exhausted way fast. So Im adding it in after and instructing the agent to use this tool inside the prompt.



```python
import os

try:
  GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
  print("Gemini API key setup complete.")
except Exception as e:
  print(f"Authetication erro: {e}")
```

```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print(" ADK components imported successfully.")
```

```python
root_agent = Agent(
  name = "helpful_assistant",
  model = "gemini-2.5-flash-lite",
  description="A simple agent that can answer general questions.",
  instruction="You are a helpful assistant.",
  tools=[]
)
print("Root Agent defined.")
```

```python
runner = InMemoryRunner(agent=root_agent)
print("Runner Created")
```

```python
response = await runner.run_debug("What is Agent Development Kid from google? What languages is the SDK available in?")
```

```python
root_agent.tools.append(google_search)
```

```python
response = await runner.run_debug("Use your Google Search tool to find the current weather in Thessaloniki, Greece.")
```
